```python
# Every one line skipped means cell change...(KNN Regression) Model1
y=model1['CGPA'].values
model1.drop(['CGPA'],axis=1,inplace=True)
#model1.drop(['Seat No.'],axis=1,inplace=True)
x=model1.values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(x,y, test_size = 0.25 , random_state=25)
```

```python
import random
def remove_str(nd_array):
    for i in nd_array:
#         new.append(i)
        for j in range(len(i)):
            if type(i[j])==str:
                y=random.uniform(0.0, 4.0)
                x=float("{0:.3f}". format(y))
                i[j] = x
                
#                 new.pop()
#                 y_arr.pop()
    return nd_array


def rounded(array):
    y_train=[]
    for i in range(len(array)):
        x=float("{0:.3f}". format(array[i]))
        y_train.append(x)
    return y_train

# a=[[1,2,'I'],[2,4,5,6],[2,4,5,6],[2,4,5,6],[2,4,5,'W']]
# print(remove_str(a))    
```

```python
y_train = rounded(y_train)
y_test = rounded(y_test)
x_train = remove_str(x_train)
x_test = remove_str(x_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors=5)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
print(y_pred)
print(y_test)
```

```python
y_pred = rounded(y_pred)
```

```python
y_pred
```

```python
sample2=[]
for i in range(1,12):
    inp=float(input('Enter GP of Course'))
    sample2.append(inp)
print(sample2)
regressor.predict([sample2])
```

```python
accuracy = r2_score(y_test,y_pred)
print("Model Accuracy =", accuracy*100, "in percent")
```
